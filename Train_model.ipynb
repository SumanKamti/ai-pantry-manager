```python
import kagglehub
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from flask import Flask, request, jsonify
import os
import shutil
import random
import tensorflow as tf
```

```python
# Download latest version
path = kagglehub.dataset_download("moltean/fruits")

print("Path to dataset files:", path)
```

```python
# Path to dataset
dataset_path = path
train_path = "fruits_split/train"
test_path = "fruits_split/test"

# Create train/test directories
for path in [train_path, test_path]:
    os.makedirs(path, exist_ok=True)

# Train-test split ratio
split_ratio = 0.8  # 80% train, 20% test

# Loop through each fruit category
for fruit in os.listdir(dataset_path):
    fruit_folder = os.path.join(dataset_path, fruit)
    if not os.path.isdir(fruit_folder):
        continue

    images = os.listdir(fruit_folder)
    random.shuffle(images)

    split_index = int(len(images) * split_ratio)
    train_images = images[:split_index]
    test_images = images[split_index:]

    # Create category folders in train/test
    os.makedirs(os.path.join(train_path, fruit), exist_ok=True)
    os.makedirs(os.path.join(test_path, fruit), exist_ok=True)

    # Move files
    for img in train_images:
        shutil.copy(os.path.join(fruit_folder, img),
                    os.path.join(train_path, fruit, img))
    for img in test_images:
        shutil.copy(os.path.join(fruit_folder, img),
                    os.path.join(test_path, fruit, img))

print("Dataset split complete!")
```
