```python
import os
import shutil
import random
```

```python

SOURCE = "dataSetML"
DEST = "data_split"

SPLIT = {"train": 0.7, "val": 0.15, "test": 0.15}

for s in SPLIT:
    os.makedirs(os.path.join(DEST, s), exist_ok=True)

for cls in os.listdir(SOURCE):
    cls_path = os.path.join(SOURCE, cls)
    if not os.path.isdir(cls_path):
        continue

    imgs = os.listdir(cls_path)
    random.shuffle(imgs)

    total = len(imgs)
    t_end = int(total * SPLIT["train"])
    v_end = t_end + int(total * SPLIT["val"])

    splits = {
        "train": imgs[:t_end],
        "val": imgs[t_end:v_end],
        "test": imgs[v_end:]
    }

    for s, files in splits.items():
        cls_dir = os.path.join(DEST, s, cls)
        os.makedirs(cls_dir, exist_ok=True)
        for f in files:
            shutil.copy(os.path.join(cls_path, f), os.path.join(cls_dir, f))

print("✅ Dataset split done")
```
